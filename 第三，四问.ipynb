```python
quality.drop(quality.index[0:2],inplace=True)#删除1-25以前的数据
quality
```

```python
def qualified(quality):
    if (quality["指标A (index A)"]>=77.78 and quality["指标B (index B)"]<=80.33 and quality["指标C (index C)"]<24.15 and quality["指标D (index D)"]<15.62):
        return 1
    else:
        return 0
```

```python
quality["合格 (qualified)"]=quality.apply(lambda quality:qualified(quality),axis=1)
quality
```

```python
quality["date"]=pd.to_datetime(pd.to_datetime(quality["时间 (Time)"],format="%m-%d-%Y").dt.strftime("%Y-%m-%d"))

'''%qtconsole取出年月日  example：date=2022-01-25.........'''
quality=quality.groupby("date").mean()
'''"一天有具体的时间点的对映不同温度，取一天所有温度的平均值，，，共74天，平均一天记录温度104029/74=1405.7972972972973次'''
```

```python
data3.dropna(inplace=True)#删除NaN值
#第四问测试数据
date=np.array([[341.40,665.04,52.88,91.2,47.22,22.26,1.25,3.09,315.613750,141.091250],
               [1010.32,874.47,54.44,92.12,48.85,21.83,1.25,3.09,281.596667,155.241111]])
x=data3[['系统I温度 (Temperature of system I)', '系统II温度 (Temperature of system II)',
       '原矿参数1 (Mineral parameter 1)',
       '原矿参数2 (Mineral parameter 2)', '原矿参数3 (Mineral parameter 3)',
       '原矿参数4 (Mineral parameter 4)', '过程数据1 (Process parameter 1)',
       '过程数据2 (Process parameter 2)', '过程数据3 (Process parameter 3)',
       '过程数据4 (Process parameter 4)']]
y=data3[['指标A (index A)', '指标B (index B)', '指标C (index C)', '指标D (index D)',
       '合格 (qualified)']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
# from sklearn.pipeline import Pipeline
# from sklearn.preprocessing import StandardScaler
# from sklearn.svm import SVC
# from sklearn.decomposition import PCA
m1=xgb.XGBRegressor()

m1.fit(x,y)
m2=LinearRegression()
m2.fit(x,y)
m3=RandomForestRegressor()
m3.fit(x,y)
m4=DecisionTreeRegressor()
m4.fit(x,y)
# m5=GradientBoostingRegressor()
# m5.fit(x,y)
m6=KNeighborsRegressor()
m6.fit(x,y)

print("xgbboost决定系数R^2:{},预测结果:{}".format(m1.score(x,y),m1.predict(date)))
print("LinearRegression决定系数R^2:{},预测结果:{}".format(m2.score(x,y),m2.predict(date)))
print("RandomForestRegressor决定系数R^2:{},预测结果:{}".format(m3.score(x,y),m3.predict(date)))
print("DecisionTreeRegressor决定系数R^2:{},预测结果:{}".format(m4.score(x,y),m4.predict(date)))
# print("GradientBoostingRegressor决定系数R^2:{},预测结果:{}".format(m5.score(x,y),m5.predict(date)))
print("KNeighborsRegressor决定系数R^2:{},预测结果:{}".format(m6.score(x,y),m6.predict(date)))
```
